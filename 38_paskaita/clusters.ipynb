```python
import numpy as np
from sklearn.cluster import AgglomerativeClustering
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.preprocessing import StandardScaler
```

```python
X = np.random.rand(150,2) * 50
# Random kintamasis
```

```python
agg_cluster = AgglomerativeClustering(n_clusters=4, linkage='complete')
agg_cluster.fit(X)
labels1 = agg_cluster.labels_
plt.scatter(X[:,0], X[:,1], c=labels1, cmap='rainbow')
plt.title('keturiu clusteriu grafikas su compleate jungimo metodu')
plt.grid()
plt.show()
```

```python
agg_cluster = AgglomerativeClustering(n_clusters=4, linkage='ward')
agg_cluster.fit(X)
labels2 = agg_cluster.labels_
plt.scatter(X[:,0], X[:,1], c=labels2, cmap='rainbow')
plt.title('keturiu clusteriu grafikas su ward jungimo metodu')
plt.grid()
plt.show()
```

```python
l, (grafikas1, grafikas2) = plt.subplots(1,2, figsize = (12, 6))
grafikas1.scatter(X[:,0], X[:,1], c = labels1)
grafikas2.scatter(X[:,0], X[:,1], c = labels2)
plt.show()

# du grafikai viename lange, palyginimui
```
