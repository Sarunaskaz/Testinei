```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

# Set up Chrome options to use incognito mode
options = Options()
options.add_argument('--incognito')

# Initialize the WebDriver with the specified options
driver = webdriver.Chrome(options=options)

# Create a list to store the product data
products = []

# Iterate through the pages you want to scrape
for page in range(2):
    url = f'https://pigu.lt/lt/mobilieji-telefonai?page={page}'
    driver.get(url)
    
    try:
        # Wait until the product links are present
        WebDriverWait(driver, 10).until(
            EC.presence_of_all_elements_located((By.XPATH, '//p[contains(@class, "product-name")]/a'))
        )
        
        # Find all product links using the corrected XPath
        nuorodos = driver.find_elements(By.XPATH, '//p[contains(@class, "product-name")]/a')
        
        # Extract the href attribute
        hrefs = [nuoroda.get_attribute('href') for nuoroda in nuorodos if nuoroda.get_attribute('href')]
        
        for href in hrefs:
            driver.get(href)
            
            # Wait until the product details are present
            WebDriverWait(driver, 10).until(
                EC.presence_of_element_located((By.CLASS_NAME, 'c-accordion-item__content'))
            )
            
            # Extract product details
            try:
                product_info = {}
                product_info['url'] = href
                product_info['model'] = driver.find_element(By.XPATH, '//tr[td[text()="Modelis:"]]/td[2]').text
                product_info['screen_size'] = driver.find_element(By.XPATH, '//tr[td[text()=" Ekrano dydis (coliais):"]]/td[2]/a').text
                product_info['main_camera'] = driver.find_element(By.XPATH, '//tr[td[text()="Pagrindinė kamera:"]]/td[2]').text
                product_info['front_camera'] = driver.find_element(By.XPATH, '//tr[td[text()="Priekinė kamera:"]]/td[2]').text
                product_info['RAM'] = driver.find_element(By.XPATH, '//tr[td[text()="Operatyvinė atmintis (RAM):"]]/td[2]').text
                product_info['battery_tipe'] = driver.find_element(By.XPATH, '//tr[td[text()="Baterijos tipas:"]]/td[2]').text
                product_info['add_funck'] = driver.find_element(By.XPATH, '//tr[td[text()="Papildomos funkcijos:"]]/td[2]').text
                product_info['proc_type'] = driver.find_element(By.XPATH, '//tr[td[text()="Procesoriaus tipas: 	"]]/td[2]').text
                
                # Append the product info to the list
                products.append(product_info)
            
            except Exception as e:
                print(f"Failed to extract product details from {href}: {e}")
    
    except Exception as e:
        print(f"Error on page {page}: {e}")


driver.quit()

df = pd.DataFrame(products)

df.to_csv('product_data.csv', index=False)

print(df)
```
