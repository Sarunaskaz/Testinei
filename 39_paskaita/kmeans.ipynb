```python
import pandas as pd 
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
data = pd.read_csv('social_network_data.csv')
```

```python
scaler = StandardScaler()
scaled_data = scaler.fit_transform(data.iloc[:, 1:]) # iloc - index location

kmeans = KMeans(n_clusters=3, random_state= 30) #random state 

clusters = kmeans.fit_predict(scaled_data)
plt.scatter(data['friends_count'], data['followers_count'], c=clusters, cmap='viridis')
plt.xlabel('draugu kiekis')
plt.ylabel('sekeju kiekis')
plt.title('Socialinio tinklo vartotoju grupes naudojant K-MEANS')
plt.show()
```

```python
customer_data = pd.read_csv('customer_data.csv')
```

```python
scaler = StandardScaler()
scaled_data = scaler.fit_transform(customer_data)

kmeans =KMeans(n_clusters=3)
clusters = kmeans.fit_predict(scaled_data)

plt.scatter(customer_data['Age'], customer_data['Income'], c=clusters, cmap='viridis')
plt.xlabel('Amzius')
plt.ylabel('pajamos')
plt.title('Klientu segmentacija pagal pajamas ir amziu')
plt.show()
```

```python
from sklearn.cluster import DBSCAN
import numpy as np
X = np.random.randn(100,2)

dbscan = DBSCAN(eps=0.3, min_samples=5)
clusters = dbscan.fit_predict(X)

plt.scatter(X[:,0], X[:,1], c=clusters, cmap='rainbow')
plt.xlabel('X1')
plt.ylabel('X2')
plt.show()
```
