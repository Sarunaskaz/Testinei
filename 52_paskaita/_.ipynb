```python
from selenium import webdriver
import os
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
import requests
options = Options()
options.add_argument('--incognito')
driver = webdriver.Chrome(options=options)
for page in range(1, 11):
    url = f'https://pigu.lt/lt/buitine-technika-elektronika/orkaites?page={page}'
    driver.get(url)
    nuorodos = driver.find_elements(By.CLASS_NAME, 'product-image-container')
    for nuoroda in nuorodos:
        href = nuoroda.get_attribute('href')
        with open('nuorodos1.txt', 'a+') as failas:
            failas.write(f'{href}\n')

## paimam nuorodas ir isauogom ja nuorodos1.txt faile
```

```python
options = Options()
options.add_argument('--incognito')
driver = webdriver.Chrome(options=options)
photos_dir = 'pvz'
if not os.path.exists(photos_dir):
    os.makedirs(photos_dir)
with open('nuorodos1.txt', 'r') as failas:
    nuorodos = failas.readlines()
# print(nuorodos)
for idx, nuoroda in enumerate(nuorodos):
    nuoroda = nuoroda.replace('\n', '')
    driver.get(nuoroda)
    nuotraukos = driver.find_elements(By.CLASS_NAME, 'c-gallery-slide--image')
    for i, nuotrauka in enumerate(nuotraukos):
        src = nuotrauka.get_attribute('src')
        img_data = requests.get(src).content
        img_name = f'{idx}_{i}.jpg'
        img_path = os.path.join(photos_dir, img_name)
        with open(img_path, 'wb') as h:
            h.write(img_data)
        print(f'Issaugota {img_name} nuotrauka')
driver.quit()

#Traukiam nuotraukas 
```

```python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
```

```python
data = "čia yra pradinis tekstas, vėliau jo sugeneruosim daugiau"
with open("tekstas.txt", "r", encoding='utf-8') as f:
  data = f.read()

raides = sorted(list(set(data)))
raides_su_index = {raide: index for index, raide in enumerate(raides)}

max_length = 40
step = 3
sakiniai = []
sekancios_raides = []

for i in range(0, len(data) - max_length, step):
  sakiniai.append(data[i:i + max_length])
  sekancios_raides.append(data[i+max_length])

x = np.zeros((len(sakiniai), max_length, len(raides)), dtype=bool)
y = np.zeros((len(sakiniai), len(raides)), dtype=bool)

# uzpildom x ir y masyvus
for i, sakinys in enumerate(sakiniai):
  for j, raide in enumerate(sakinys):
    x[i,j,raides_su_index[raide]] = 1
  y[i, raides_su_index[sekancios_raides[i]]] = 1

model = Sequential([ 
    LSTM(128, input_shape=(max_length, len(raides))),
    Dense(len(raides), activation='softmax')
])

model.compile(optimizer='adam', loss='categorical_crossentropy')

model.fit(x, y, batch_size=128, epochs=30)
```

```python
sugeneruotas_tekstas = "Gyvenimo prasmė yra "

for _ in range(100):
  x_pred = np.zeros((1, max_length, len(raides)))
  for i, raide in enumerate(sugeneruotas_tekstas[-max_length:]):
    x_pred[0, i, raides_su_index[raide]] = 1

  y_pred = model.predict(x_pred, verbose=0)[0]
  next_index = np.random.choice(len(raides), p=y_pred)
  next_raide = raides[next_index]

  sugeneruotas_tekstas += next_raide

print(sugeneruotas_tekstas)
```
